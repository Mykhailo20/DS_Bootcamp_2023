```python
# first neural network with keras tutorial
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt
```

```python
def plot_graph(history, string):
    plt.plot(history.history[string])
    plt.xlabel('Epochs')
    plt.ylabel(string)
    plt.show()
```

```python
# load the dataset
dataset = loadtxt('data/pima-indians-diabetes.csv', delimiter=',')
# split into input (X) and output (y) variables
X = dataset[:, 0:8]
y = dataset[:, 8]
```

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_shape=(8, ), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
history = model.fit(X, y, epochs = 150, batch_size=10)
```

```python
plot_graph(history, 'accuracy')
plot_graph(history, 'loss')
```

## Evaluate keras model

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

## Make predictions

#### You can adapt the above example and use it to generate predictions on the training dataset, pretending it is a new dataset you have not seen before.

```python
# make class predictions with the model
predictions = (model.predict(X) > 0.5).astype(int)

# summarise the first 5 cases
for i in range(5):
    print(f"{X[i].tolist()} => {predictions[i][0]} (true value = {y[i]})")
```
